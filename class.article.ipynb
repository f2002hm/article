```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time

# =============================
#  Truck
# =============================
class Truck:
    def __init__(self, truck_id, data_row=None):
        self.id = truck_id
        self.data_row = data_row  #   (  )
        self.unload_time = np.array([])
        self.loading_time = np.array([])
        self.total_cycle_time = np.array([])
        self.travel_time = np.array([])
        self.failure_prob = np.array([])
        self.failure_time = np.array([])

    def generate_pert_data(self, a, m, b, n_sim=1000):
        alpha = 1 + 4 * (m - a) / (b - a)
        beta_param = 1 + 4 * (b - m) / (b - a)
        return a + (b - a) * np.random.beta(alpha, beta_param, size=n_sim)

    def prepare_profile(self):
        #   PERT     
        self.loading_time = self.generate_pert_data(3, 4.5, 6)
        self.unload_time = self.generate_pert_data(1, 2, 3)
        self.total_cycle_time = self.generate_pert_data(20, 23, 27)
        self.travel_time = (self.total_cycle_time - self.loading_time - self.unload_time) / 2
        #   
        self.failure_prob = np.random.rand(len(self.total_cycle_time))
        self.failure_time = self.failure_prob * self.total_cycle_time

    def profile(self):
        return {
            "loading_time": self.loading_time,
            "unload_time": self.unload_time,
            "travel_time": self.travel_time,
            "failure_prob": self.failure_prob,
            "failure_time": self.failure_time
        }

# =============================
#  Simulation
# =============================
class Simulation:
    @staticmethod
    def monte_carlo(data, n_sim=10000):
        start = time.time()
        samples = np.random.choice(data, size=n_sim, replace=True)
        return samples, time.time() - start

    @staticmethod
    def pert_params(data):
        a, b = np.min(data), np.max(data)
        m = np.mean(data)
        return a, m, b

    @staticmethod
    def pert(data_or_params, n_sim=10000):
        if isinstance(data_or_params, (list, np.ndarray, pd.Series)):
            a, m, b = Simulation.pert_params(data_or_params)
        else:
            a, m, b = data_or_params
        start = time.time()
        alpha = 1 + 4 * (m - a) / (b - a)
        beta_param = 1 + 4 * (b - m) / (b - a)
        samples = a + (b - a) * np.random.beta(alpha, beta_param, size=n_sim)
        return samples, time.time() - start

# =============================
#  Visualization
# =============================
class Visualization:
    @staticmethod
    def plot_mc_vs_pert(data, title):
        mc, _ = Simulation.monte_carlo(data)
        a, m, b = Simulation.pert_params(data)
        pert, _ = Simulation.pert((a, m, b))

        plt.figure()
        plt.hist(mc, bins=30, density=True, alpha=0.5, label="Monte Carlo")
        plt.hist(pert, bins=30, density=True, alpha=0.5, label="PERT")
        plt.legend()
        plt.title(title)
        plt.xlabel("Value")
        plt.ylabel("Probability Density")
        plt.show()

# =============================
#  Analysis
# =============================
class Analysis:
    @staticmethod
    def comparison_table(data):
        mc, mc_time = Simulation.monte_carlo(data)
        a, m, b = Simulation.pert_params(data)
        pert, pert_time = Simulation.pert((a, m, b))

        table = pd.DataFrame({
            "Method": ["Monte Carlo", "PERT"],
            "Mean": [mc.mean(), pert.mean()],
            "Variance": [mc.var(), pert.var()],
            "Execution Time (s)": [mc_time, pert_time]
        })
        return table

    @staticmethod
    def analyze_trucks(truck_list):
        for truck in truck_list:
            print(f"\n===== Truck {truck.id} Analysis =====")
            profile = truck.profile()
            for key in ["loading_time", "travel_time", "unload_time", "failure_time"]:
                print(f"\n{key.replace('_', ' ').title()} Comparison")
                print(Analysis.comparison_table(profile[key]))
                Visualization.plot_mc_vs_pert(profile[key], f"Truck {truck.id} - {key.replace('_', ' ').title()}: MC vs PERT")

# =============================
#       
# =============================
if __name__ == "__main__":
    #    
    df = pd.read_excel("Book.xlsx")  #   =  
    trucks = []

    for i, row in df.iterrows():
        truck = Truck(truck_id=i+1, data_row=row)
        truck.prepare_profile()
        trucks.append(truck)

    #   
    Analysis.analyze_trucks(trucks)
```
