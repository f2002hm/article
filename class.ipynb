```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time

# =============================
# Truck
# =============================
class Truck:
    def __init__(self, truck_id):
        self.id = truck_id
        #    30 ( )
        self.loading_time = np.zeros(30)
        self.unload_time = np.zeros(30)
        self.total_cycle_time = np.zeros(30)
        self.travel_time = np.zeros(30)
        self.failure_time = np.zeros(30)

    @staticmethod
    def generate_pert_data(a, m, b, n=30):
        alpha = 1 + 4 * (m - a) / (b - a)
        beta_param = 1 + 4 * (b - m) / (b - a)
        return a + (b - a) * np.random.beta(alpha, beta_param, size=n)

    def prepare_profile(self):
        n = 30  #  
        self.loading_time = self.generate_pert_data(3, 4.5, 6, n)
        self.unload_time = self.generate_pert_data(1, 2, 3, n)
        self.total_cycle_time = self.generate_pert_data(20, 23, 27, n)
        self.travel_time = (self.total_cycle_time - self.loading_time - self.unload_time) / 2
        self.failure_time = np.random.rand(n) * self.total_cycle_time

    def profile(self):
        return {
            "loading_time": self.loading_time,
            "unload_time": self.unload_time,
            "travel_time": self.travel_time,
            "failure_time": self.failure_time
        }

# =============================
# Simulation
# =============================
class Simulation:
    @staticmethod
    def monte_carlo(data, n_sim=10000):
        start = time.time()
        samples = np.random.choice(data, size=n_sim, replace=True)
        return samples, time.time() - start

    @staticmethod
    def pert_from_data(data, n_sim=10000):
        a, b = data.min(), data.max()
        m = data.mean()
        alpha = 1 + 4 * (m - a) / (b - a)
        beta_param = 1 + 4 * (b - m) / (b - a)
        start = time.time()
        samples = a + (b - a) * np.random.beta(alpha, beta_param, size=n_sim)
        return samples, time.time() - start

# =============================
# Visualization
# =============================
class Visualization:
    @staticmethod
    def plot_mc_vs_pert(data, title):
        if len(data) == 0:
            print(f"No data to plot for {title}")
            return
        mc, _ = Simulation.monte_carlo(data)
        pert, _ = Simulation.pert_from_data(data)

        plt.figure()
        plt.hist(mc, bins=30, density=True, alpha=0.5, label="Monte Carlo")
        plt.hist(pert, bins=30, density=True, alpha=0.5, label="PERT")
        plt.legend()
        plt.title(title)
        plt.xlabel("Value")
        plt.ylabel("Density")
        plt.show()

# =============================
# Analysis
# =============================
class Analysis:
    @staticmethod
    def comparison_table(data):
        if len(data) == 0:
            return pd.DataFrame()
        mc, mc_t = Simulation.monte_carlo(data)
        pert, pert_t = Simulation.pert_from_data(data)

        return pd.DataFrame({
            "Method": ["Monte Carlo", "PERT"],
            "Mean": [mc.mean(), pert.mean()],
            "Variance": [mc.var(), pert.var()],
            "Execution Time (s)": [mc_t, pert_t]
        })

    @staticmethod
    def analyze_trucks(truck_list):
        for truck in truck_list:
            print(f"\n===== Truck {truck.id} Analysis =====")
            profile = truck.profile()
            for key in ["loading_time", "unload_time", "travel_time", "failure_time"]:
                if key in profile and len(profile[key]) > 0:
                    print(f"\n{key.replace('_', ' ').title()} Comparison")
                    print(Analysis.comparison_table(profile[key]))
                    Visualization.plot_mc_vs_pert(profile[key], f"Truck {truck.id} - {key.replace('_', ' ').title()}: MC vs PERT")

# =============================
# Run
# =============================
if __name__ == "__main__":
    #   
    trucks = [Truck(i+1) for i in range(10)]
    for truck in trucks:
        truck.prepare_profile()

    Analysis.analyze_trucks(trucks)
```
